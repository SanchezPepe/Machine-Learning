```python
import pandas as pnd
import numpy as np
import random as rd
import matplotlib.pyplot as plot
```

```python
sample_size = 100
data1 = np.random.normal(-40,25, size = (sample_size, 2))
data2 = np.random.normal(80,25, size = (sample_size, 2))
type1 = np.ones((sample_size,1))
type2 = np.ones((sample_size,1)) * -1

class1 = np.hstack((data1, type1))
class2 = np.hstack((data2, type2))
data = np.vstack((class1,class2))
```

```python
plot.plot(data1[:, 0], data1[:, 1], 'g^', color= 'g')
plot.plot(data2[:, 0], data2[:, 1], 'o', color= 'b')
plot.show()
```

```python
def perceptron(data, T):
    X = data[:, :-1] 
    Y = data[:, -1]
    n = len(X)
    
    theta = np.zeros(shape=(1, X.shape[1])) 
    theta0 = 0
    
    for t in range(T):
        for i in range(n):
            if np.dot(Y[i], np.dot(theta, X[i]) + theta0) <= 0:
                theta += np.dot(Y[i],X[i])
                theta0 += Y[i]

    return theta, theta0
```

```python
theta, theta0 = perceptron(data, 1000)
theta, theta0
```

```python
x = np.linspace(data[:, 0].min(), data[:, 0].max(), 10)
y = (-x * theta[0, 0] - theta0) / theta[0,1]

plot.plot(x,y,'r', color = 'r')
plot.plot(data1[:, 0], data1[:, 1], 'g^', color= 'g')
plot.plot(data2[:, 0], data2[:, 1], 'o', color= 'b')
plot.show()
```
